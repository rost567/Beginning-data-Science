Now using data about the average height in countries around the world. It comes from Kaggle.

```python
height = pd.read_csv("height_2022.csv")
height = height.drop(["Male Height in Ft", "Female Height in Ft"], axis=1)
height.head()
```

Now manipulating the dataframe "height" to be tidy. Currently, multiple variables are stored in one column.
> - it has a new column "Sex" containing a single character "M" or "F" for "male"/"female".
> - it has a new column "Height in Cm" containing data about the height in cm

```python
height_tidy = pd.melt(height, id_vars=['Rank', 'Country Name'], var_name='Sex', value_name='Height in Cm')
height_tidy['Sex'] = height_tidy['Sex'].replace(['Male Height in Cm','Female Height in Cm'],['M','F']) 
 
height_tidy.head()
```

Now working with data about billboards in the 2000s.

```python
billboard = pd.read_csv("billboard_intermediate.csv")
billboard.head(5)
```

Manipulating the dataframe "billboard" into two dataframes that each store one observational unit.

> - Making a new dataframe "billboard_songs" that contains information about the songs. It should contain these columns from the "billboard" dataframe: "artist.inverted", "track", "time", "genre" 
> - Making a new dataframe "billboard_charts" that contains information about the chart performance of the songs
It should contain these columns from the "billboard" dataframe: "date", "date.peaked", "week", "rank".
> - Additionally, both dataframes should contain a column named "track_id" that gives each track a unique ID and links the information in the two dataframes.

```python
billboard['track_id'] = billboard.groupby('track').ngroup()
billboard_songs = billboard[['artist.inverted','track','time','genre','track_id']]
billboard_songs = billboard_songs.groupby(['artist.inverted', 'time', 'genre','track'], sort=False, as_index=False)['track_id'].first()

billboard_songs.head()
```

```python
billboard_charts = billboard[['date', 'date.peaked', 'week', 'rank','track_id']]
billboard_charts.head()
```

What was the number one hit (rank = 1) on 2000/07/29? This question may seem random (and it is), but this is the 2000s! This is the day on which Jennifer Aniston married Brad Pitt. Did they play this song? Who knows.
> Define one variable
"name_song": Name of the song that was at rank 1 on the given date

```python
name_song = billboard.loc[(billboard["date.peaked"] == "2000-07-29") & (billboard["rank"] == 1.0),['track']]
name_song = name_song.groupby(['track'], sort=False, as_index=False).first()
name_song = name_song.iloc[0]['track']
name_song
```

About average heights. 
What's the country with the tallest men?
Defining two variables "country_tallest_men" as a string and "height_tallest_men".

```python
country_tallest_men_ = height_tidy.loc[(height_tidy['Sex']== 'M')]
country_tallest_men = country_tallest_men_.iloc[0]['Country Name']
country_tallest_men

height_tallest_men = country_tallest_men_.iloc[0]['Height in Cm']
height_tallest_men
```

In which country is the difference of average heights between men and women largest? 


```python
height['height difference'] = height['Male Height in Cm'].astype('float') - height['Female Height in Cm'].astype('float')
largest_difference = height.sort_values(by='height difference', ascending=False)

country_largest_difference = largest_difference.iloc[0]['Country Name']

largest_difference = largest_difference.iloc[0]['height difference']

country_largest_difference
largest_difference
```

The Pearson's correlation value of average male and average female height.

```python
corr_avg_heights = height['Male Height in Cm'].corr(height['Female Height in Cm'])
corr_avg_heights
```

Plotting the basic information to get a sense of the dataset.

Does the unemployment rate look proportional to the rank of the major? To see this, we will display the unemployment rate against the rank of the major.

```python
df.head()
df.columns
df.columns = list(map(str, df.columns))
plt.plot(df['Rank'], df['Unemployment_rate'])

plt.xlabel('Rank')
plt.ylabel('Unemployment_rate')

plt.show()
```

Does the salary look proportional to the rank of the major? 
For this we will display the median salary, the 25th and 75th percentiles against the rank of the major. 
We will plot three lines in the same plot.

```python
plt.figure(figsize=(12,8))
plt.title('', fontdict={'fontweight': 'bold', 'fontsize':22})

columns_to_look_at = ['Median', 'P25th', 'P75th']
for  columns in df:
    if columns in columns_to_look_at: 
        plt.plot(df.Rank, df[columns], marker='.', label = columns)


plt.xlabel('Rank')
plt.ylabel('Salary against the rank')
plt.legend()
plt.show()
```

Now analyzing and visualizing the categorical data.  

We want to count how many students there are per category.
Extracting the different categories (column "Major_category") and displaying them on an horizontal bar plot.
Summing up the number of students in each category (column "Total").


```python
cat_totals = df.groupby(['Major_category'])['Total'].sum()
plt.figure(figsize=(20,12))
cat_totals.plot(kind='barh', legend=False, colormap='Spectral')
plt.xlabel('Total of students')
plt.show()
```

```python
assert_equal(len(cat_totals),16)
assert_equal(cat_totals["Interdisciplinary"],12296)
assert_equal(cat_totals["Engineering"],537583)
```

- Vertical and horizontal bar charts are often a good choice if you want to see the difference between your categories.  
- If the interest is in ratios, then pie plots are a good tool.  
- However, since cat_totals contains a few smaller categories, creating a pie plot with cat_totals.plot(kind="pie") will produce several tiny slices with overlapping labels.

```python
cat_totals.plot(kind="pie", label='')
plt.show()
```

To solve this problem, we have to merge the smaller categories into a single group.  
> - First, separating the data in two groups, depending on the number of occurence. 
> - The small group should contain all major category that have less than 200,000 individual.
> - Making sure that the variable containing all the small groups is named something like "small_....".

```python
small_cat_totals = cat_totals[cat_totals<200000]
big_cat_totals = cat_totals[cat_totals>=200000]
```

```python
small_sums = small_cat_totals.sum()
small_sums = pd.Series(small_sums, index=['Other'])
```

```python
assert_equal(small_cat_totals["Agriculture & Natural Resources"], 75620)
assert_equal(small_cat_totals["Interdisciplinary"], 12296)

assert_equal(small_cat_totals["Law & Public Policy"],179107)
assert_equal(small_cat_totals["Physical Sciences"], 185479)
```

```python
big_cat_totals = pd.concat([big_cat_totals, small_sums], copy = False)
#big_cat_totals

plt.figure(figsize=(20,12))
big_cat_totals.plot(kind='pie',autopct='%.2f %%', colormap='Spectral', label='')

plt.figure(figsize=(20,12))
plt.show()
```
